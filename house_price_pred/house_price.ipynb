```python
import pandas as pd
```

```python
df = pd.read_csv("/content/town.csv")
```

```python
df.head()
```

```python
pd.get_dummies(df.town,dtype=int)
```

```python
dummies = pd.get_dummies(df.town,dtype=int)
```

```python
merged = pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
final = merged.drop(['town'],axis='columns')
```

```python
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop(['price'], axis='columns')
y = final.price
```

```python
model.fit(x,y)
```

```python
y
```

```python
model.predict([[2800,1,0,0]]
)
```

```python
model.score(x,y)
```
